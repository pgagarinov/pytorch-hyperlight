```python
from pytorch_hyperlight.datasets.cifar import CIFARDataLoadersBuilder
from pytorch_hyperlight.utils.image_utils import show_image_tensors
```

```python
def basic_usage(builder):
    f_create_dataloaders, f_create_datasets = builder.build()

    BATCH_SIZE = 20

    loaders_dict = f_create_dataloaders(batch_size=BATCH_SIZE)

    batch = next(iter(loaders_dict['train_show_full_augmented_loader']))

    assert batch[0].shape[0] == BATCH_SIZE
    assert batch[1].shape[0] == BATCH_SIZE

    show_image_tensors(list(batch[0]), max_cols=5)
```

```python
builder = CIFARDataLoadersBuilder(dataset_name='cifar100')
```

```python
basic_usage(builder)
```

```python
from pytorch_hyperlight.datasets.kaggle.utkfaces import UTKFacesDataLoadersBuilder
```

```python
builder = UTKFacesDataLoadersBuilder(label_field='age')
basic_usage(builder)
```
